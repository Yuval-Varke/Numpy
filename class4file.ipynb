##Fancy Indexing
you can Pick elements from specific positions using a list/array of indices.

```python
import numpy as np
a = np.array([1,2,3,4,5,6,7])
a[[1,5,3]]
```

You can do this fancy indexing in 2D Matrix aswell

```python
mat = np.arange(1,13).reshape(3,4)
print(mat)
mat[[0,1,2],[1,2,3]]
```

##Conditional Selection
Find values based on conditions or replace them.

```python
b = np.array([1,2,3,4,5,6,7])
# print(b[b>2])
np.where(b>3,b,0)
```

##Sorting & Ordering
np.sort() → returns sorted array

np.argsort() → returns indices of sorted order

```python
c = np.array([6,3,4,1,0,-2])
np.sort(c)
```

```python
np.argsort(c)
```

##Concatenation & Stacking

```python
x = np.array([1,2,3])
y = np.array([4,6,7])
np.concat([x,y])
```

Vertical & Horizontal Stacking

```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
```

```python
np.vstack((A,B))
```

```python
np.hstack((B,A))
```

```python
np.concat([A,B])
```
